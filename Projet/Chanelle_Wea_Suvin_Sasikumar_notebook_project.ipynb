Function to find features that a highly positive or negative correlated to other features

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr

corr_features1 = correlation(year1, 0.60)
```

## Data Modeling : Prediction Model for columns X65

```python
X = year1.iloc[: ,:-1]
```

```python
Y = year1.iloc[:,-1]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.33)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

# 1st Model : Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 5, criterion = 'entropy')
rf.fit(X_train,Y_train)
```

```python
y_pred=rf.predict(X_test)
```

```python
from sklearn import metrics
a =metrics.accuracy_score(Y_test, y_pred)
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred))
```

# 2nd Model : Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression(solver='liblinear', random_state=0)
lr.fit(X_train,Y_train)
y_pred_lr=lr.predict(X_test)
b=metrics.accuracy_score(Y_test, y_pred_lr)
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred_lr))
```

# 3rd Model : Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,Y_train)
y_pred_clf = clf.predict(X_test)
c=metrics.accuracy_score(Y_test, y_pred_clf)
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred_clf))
```

# 4th Model : Bagging

```python
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
clfbg = BaggingClassifier(base_estimator=SVC(),n_estimators=10, random_state=0)
clfbg.fit(X_train, Y_train)
y_pred_bg = clfbg.predict(X_test)
d=metrics.accuracy_score(Y_test, y_pred_bg)
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred_bg))
```

# 5th Model : Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
clfb = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0)
clfb.fit(X_train, Y_train)
y_pred_b = clfb.predict(X_test)
f=metrics.accuracy_score(Y_test, y_pred_b)
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred_b))
```

# 6th Model : KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)
```

```python
knn.fit(X_train, Y_train)
y_pred_knn = knn.predict(X_test)
```

```python
g=metrics.accuracy_score(Y_test, y_pred_knn)
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred_knn))
```
